```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
file='Williamsburg-Bibs-for-Website-.csv'

data=pd.read_csv(file, error_bad_lines=False,index_col=2)
data=data[data['Bib'].str.contains("Relay") == False]
data['Age']=data['Age'].astype(int)
```

```python
data.groupby('Gender').Age.describe()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,6))

for label, df in data.groupby('Gender'):
    df.Age.plot(ax=axes[0],kind="hist",alpha=0.5,bins=np.arange(15,75,5),label=label,grid=True).legend()
for label, df in data.groupby('Gender'):
    df.Age.plot(ax=axes[1],kind="kde", xlim=[15,75], label=label,grid=True).legend()
axes[0].set_title('Histogram');
axes[1].set_title('Density');
axes[0].set_xlabel('Age');
axes[1].set_xlabel('Age');

fig.suptitle('Participants Age by Gender [bsan.eu/rev3w-p16]')
fig.set_facecolor('white');
```
