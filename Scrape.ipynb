```python
url = 'http://timing.rev3tri.com/mobile/athlete-results/D5520FB7-20E3-407B-9A11-A666E43BA256/2/1235'
def parse_racer_page( url ):
    "Given a Rev3 results page, scrape variables into a dictionary"
    r = requests.get(url)
    # Turn the HTML into a Beautiful Soup object
    soup = BeautifulSoup(r.text, 'lxml') 
    #get variables
    racer={}
    racer['name']=soup.find_all('h2')[1].string.strip()
    #overview table
    table=soup.find_all('table')[0]
    rows=table.find_all('tr')
    racer['bib']=int(rows[0].find_all('td')[1].string.strip())
    racer['division']=rows[1].find_all('td')[1].string.strip()
    racer['gender']=rows[2].find_all('td')[1].string.strip()
    racer['total_time']=rows[4].find_all('td')[1].string.strip().split(".")[0]
    #overview table
    table=soup.find_all('table')[1]
    rows=table.find_all('tr')[1:]
    racer['swim_time']=rows[0].find_all('td')[1].string.strip().split(".")[0]
    racer['swim_pace']=rows[0].find_all('td')[2].string.strip().split(" ")[0]
    racer['bike_time']=rows[1].find_all('td')[1].string.strip().split(".")[0]
    racer['bike_pace']=float(rows[1].find_all('td')[2].string.strip().split(" ")[0])
    racer['run_time']=rows[2].find_all('td')[1].string.strip().split(".")[0]
    racer['run_pace']=rows[2].find_all('td')[2].string.strip().split(" ")[0]
    #split table
    table=soup.find_all('table')[2]
    rows=table.find_all('tr')[1:]
    racer['division']=rows[0].find_all('td')[0].string.strip()
    racer['swim_pace']=rows[1].find_all('td')[3].string.strip().split(" ")[0]
    racer['swim_r_division']=int(rows[1].find_all('td')[4].string.strip().split(" ")[0])
    racer['swim_r_gender']=int(rows[1].find_all('td')[5].string.strip().split(" ")[0])
    racer['swim_r_overall']=int(rows[1].find_all('td')[6].string.strip().split(" ")[0])
    racer['t1_time']=rows[2].find_all('td')[2].string.strip()
    racer['bike_sA31_time']=rows[4].find_all('td')[2].string.strip()
    racer['bike_sA31_speed']=rows[4].find_all('td')[3].string.strip().split(" ")[0]
    racer['bike_sA31_r_division']=int(rows[4].find_all('td')[4].string.strip().split(" ")[0])
    racer['bike_sA31_r_gender']=int(rows[4].find_all('td')[5].string.strip().split(" ")[0])
    racer['bike_sA31_r_overall']=int(rows[4].find_all('td')[6].string.strip().split(" ")[0])
    racer['bike_sB25_time']=rows[5].find_all('td')[2].string.strip()
    racer['bike_sB25_speed']=rows[5].find_all('td')[3].string.strip().split(" ")[0]
    racer['bike_sB25_r_division']=int(rows[5].find_all('td')[4].string.strip().split(" ")[0])
    racer['bike_sB25_r_gender']=int(rows[5].find_all('td')[5].string.strip().split(" ")[0])
    racer['bike_sB25_r_overall']=int(rows[5].find_all('td')[6].string.strip().split(" ")[0])
    racer['t2_time']=rows[6].find_all('td')[2].string.strip()
    racer['run_sA2.75_time']=rows[8].find_all('td')[2].string.strip()
    racer['run_sA2.75_speed']=rows[8].find_all('td')[3].string.strip().split(" ")[0]
    racer['run_sA2.75_r_division']=int(rows[8].find_all('td')[4].string.strip().split(" ")[0])
    racer['run_sA2.75_r_gender']=int(rows[8].find_all('td')[5].string.strip().split(" ")[0])
    racer['run_sA2.75_r_overall']=int(rows[8].find_all('td')[6].string.strip().split(" ")[0])
    racer['run_sB9.05_time']=rows[9].find_all('td')[2].string.strip()
    racer['run_sB9.05_speed']=rows[9].find_all('td')[3].string.strip().split(" ")[0]
    racer['run_sB9.05_r_division']=int(rows[9].find_all('td')[4].string.strip().split(" ")[0])
    racer['run_sB9.05_r_gender']=int(rows[9].find_all('td')[5].string.strip().split(" ")[0])
    racer['run_sB9.05_r_overall']=int(rows[9].find_all('td')[6].string.strip().split(" ")[0])
    racer['run_sC4.05_time']=rows[10].find_all('td')[2].string.strip()
    racer['run_sC4.05_speed']=rows[10].find_all('td')[3].string.strip().split(" ")[0]
    racer['run_sC4.05_r_division']=int(rows[10].find_all('td')[4].string.strip().split(" ")[0])
    racer['run_sC4.05_r_gender']=int(rows[10].find_all('td')[5].string.strip().split(" ")[0])
    racer['run_sC4.05_r_overall']=int(rows[10].find_all('td')[6].string.strip().split(" ")[0])
    return racer
parse_racer_page(url)
```

```python
url='http://timing.rev3tri.com/mobile/athlete-search/D5520FB7-20E3-407B-9A11-A666E43BA256/2'
domain="http://timing.rev3tri.com"

#collect urls of racers
"Given a Rev3 results page, scrape variables into a dictionary"
racers_urls=[]
for page in np.arange(1,16):
    r = requests.get(url+'/'+str(page))
    soup = BeautifulSoup(r.text, 'lxml') 
    print(url+'/'+str(page),'-',len(soup.find_all('table')[1].find_all("a")))
    for link in soup.find_all('table')[1].find_all("a")[4:]:
        racers_urls.append(domain+link.get("href"))
racers_urls
```

```python
runner={}
for url in racers_urls:
    try:
        racer=parse_racer_page( url )
        racer['error']=''
    except:
        pass
        racer['error']='yes'
    runner[racer['bib']]=racer
    print(racer['bib'],racer['error'])
```

```python
df = pd.DataFrame(runner).transpose()
```
